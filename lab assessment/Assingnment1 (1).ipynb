```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = x**2 + 3*x + 2

plt.plot(x, y)
plt.title('Plot of y = x^2 + 3x + 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

min_index = np.argmin(y)
x_min = x[min_index]
y_min = y[min_index]
print(f"Minimum point: ({x_min:.2f}, {y_min:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 4, 100)
y = np.log(x)

plt.plot(x, y)
plt.title('Logarithmic Function: y = log(x)')
plt.xlabel('x')
plt.ylabel('y = log(x)')
plt.grid(True)
plt.show()

min_index = np.argmin(y)
x_min = x[min_index]
y_min = y[min_index]
print(f"Minimum point: (x = {x_min:.2f}, y = {y_min:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x) + np.cos(2*x)

plt.plot(x, y)
plt.title('Plot of y = sin(x) + cos(2x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

min_index = np.argmin(y)
x_min = x[min_index]
y_min = y[min_index]
print(f"Minimum point: ({x_min:.2f}, {y_min:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 100)


X, Y = np.meshgrid(x, y)
Z = np.cos(X) * np.sin(Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('z = cos(x) * sin(y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()

min_indices = np.unravel_index(np.argmin(Z), Z.shape)
x_min = X[min_indices]
y_min = Y[min_indices]
z_min = Z[min_indices]
print(f"Minimum point: ({x_min:.2f}, {y_min:.2f}, {z_min:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = X**2 / 4 + Y**2 / 9

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title(r'$z = \frac{x^2}{4} + \frac{y^2}{9}$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()

z_min = np.min(Z)
min_indices = np.unravel_index(np.argmin(Z), Z.shape)

x_min = X[min_indices]
y_min = Y[min_indices]
print(f"Minimum point: (x = {x_min:.2f}, y = {y_min:.2f}, z = {z_min:.2f})")
```
