```python
#Question 1
import math
def cal_f(x):
    return 4 * x - 1.8 * x ** 2 + 1.2 * x ** 3 - 0.3 * x ** 4
def cal_d(xu, xl):
    R = (math.sqrt(5) - 1) * 0.5
    return R * (xu - xl)
def check_e(xu, xl, x_opt, e):
    R = (math.sqrt(5) - 1) * 0.5
    ea = (1 - R) * abs((xu - xl) / x_opt)
    print("Ea = {} %".format(ea * 100))
    return ea < e
def search(xu, xl, e, n):
    x = [0, 0]
    f = [0, 0]
    result = 0
    for i in range(n):
        d = cal_d(xu, xl)
        x[0] = xl + d
        x[1] = xu - d
        f[0] = cal_f(x[0])
        f[1] = cal_f(x[1])
        if f[0] > f[1]:
            xl = x[1]
            result = x[0]
            if check_e(xu, xl, x[0], e):
                break
        else:
            xu = x[0]
            result = x[1]
            if check_e(xu, xl, x[1], e):
                break
        if i % 1 == 0:
            print("Iteration {}: xL = {}\t xU = {}\t x1 = {}\t fx1 = {}\t x2 = {}\t fx2 = {}\t d = {}\t"
                  .format(i, round(xl, 5), round(xu, 5), round(x[0], 5), round(f[0], 5), round(x[1], 5),
                          round(f[1], 5), round(d, 5)))
    return result
xU = 4
xL = 2
eS = 1 / 100
n = 100
result = search(xU, xL, eS, n)
print("x = {}, fx = {}".format(result, cal_f(result)))
```
