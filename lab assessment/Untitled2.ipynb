```python
from scipy.integrate import odeint
import numpy as  np
from matplotlib import pyplot as plt
def f(y,x):
    return x
y0 = 1
xs = np.arange(0,1,0.1)
ys = odeint(f,y0,xs)
plt.plot(xs,ys,'-')
plt.plot(xs,ys,'ro')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title('differential equation:dy/dx=x,y(0)=1,xrange=[0,1]')
plt.show
```

```python
from scipy.integrate import odeint
import numpy as  np
from matplotlib import pyplot as plt
def f(y,x):
    return(x*y/np.sqrt(y**2+x))
y0= 1
xs = np.arange(0,5,0.25)
ys = odeint(f,y0,xs)
plt.plot(xs,ys,'-')
plt.plot(xs,ys,'ro')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title('Differential equation dy/dy=(x*y/(np.sqrt(y**2+x))')
plt.show
```

```python
from scipy.integrate import odeint
import numpy as  np
from matplotlib import pyplot as plt
def f(u,t):
    dudt =(1-2*t)*u
    return dudt
u0 = 1
xs = np.arange(-1,2,0.2)
ys = odeint(f,u0,xs)
plt.plot(xs,ys,'-')
plt.plot(xs,ys,'ro')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title('differential equation:du/dx=(1-2*t)*t,xrange=[0,1]')
plt.show
```

```python
from scipy.integrate import odeint
import numpy as  np
from matplotlib import pyplot as plt
def f(y,x):
    dydx =(y/x)
    return dydx
y1 = 1
xs = np.arange(1,10,1)
ys = odeint(f,y1,xs)
plt.plot(xs,ys,'-')
plt.plot(xs,ys,'ro')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title('differential equation:dy/dx=(y/x),xrange=[1,10,1]')
plt.show
```
