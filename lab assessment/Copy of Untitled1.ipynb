```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
y = x**2 + 2*x + 2

plt.plot(x,y)
plt.title('plot y = x^2 + 2x + 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

min_index = np.argmin(y)
x_min = x[min_index]
y_min = y[min_index]
print(f"minimum point: ({x_min:.2f}, {y_min:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.05,6,100)
y = np.log(x)

plt.plot(x,y)
plt.title('logrithmic function: y = log(x)')
plt.xlabel('x')
plt.ylabel('y = log(x)')
plt.grid(True)
plt.show()

min_index =np.argmin(y)
x_min = x[min_index]
y_min = y[min_index]
print(f"minimum point: ({x_min:.3f}, {y_min:.3f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-1*np.pi, 1*np.pi, 100)
y = np.linspace(-1*np.pi, 1*np.pi, 100)


X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('z = sin(x) * cos(y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()

min_indices = np.unravel_index(np.argmin(Z), Z.shape)
x_min = X[min_indices]
y_min = Y[min_indices]
z_min = Z[min_indices]
print(f"Minimum point: ({x_min:.3f}, {y_min:.3f}, {z_min:.3f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi, 100)
y = np.cos(4*x) + np.sin(5*x)

plt.plot(x, y)
plt.title('Plot of y = cos(4x) + sin(5x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

min_index = np.argmin(y)
x_min = x[min_index]
y_min = y[min_index]
print(f"Minimum point: ({x_min:.3f}, {y_min:.3f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-3, 10, 100)
y = np.linspace(-3, 10, 100)

X, Y = np.meshgrid(x, y)
Z = X**3 / 6 + Y**3 / 12

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title(r'$z = \frac{x^3}{6} + \frac{y^3}{12}$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()

z_min = np.min(Z)
min_indices = np.unravel_index(np.argmin(Z), Z.shape)

x_min = X[min_indices]
y_min = Y[min_indices]
print(f"Minimum point: (x = {x_min:.3f}, y = {y_min:.3f}, z = {z_min:.3f})")
```
