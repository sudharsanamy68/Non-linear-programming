```python
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

def f(x1, x2):

    return 100 * (x2 - x1**2)**2 + (1 - x1)**2

def generalized_fibonacci_search(f, a, b, tol=1e-6, max_iter=6):

    # Initialize Fibonacci numbers
    fib_prev, fib_curr = 1, 1

    # Compute the total length of the interval
    L = b - a

    # Iterate until convergence or max iterations
    for _ in range(max_iter):
        # Compute the ratio
        r = fib_prev / fib_curr

        # Update the interval endpoints
        x1 = a + (1 - r) * L
        x2 = a + r * L

        # Evaluate the function at the new points
        f1 = f(x1, x2)
        f2 = f(x1, x2)

        # Update the interval
        if f1 < f2:
            b = x2
        else:
            a = x1

        # Update Fibonacci numbers
        fib_prev, fib_curr = fib_curr, fib_prev + fib_curr

        # Check convergence
        if abs(b - a) < tol:
            break

    # Return the midpoint of the final interval
    return (a + b) / 2

# Example usage:
a, b = 0, 0.1
min_x = generalized_fibonacci_search(f, a, b)
min_value = f(min_x, min_x)


print(f"Minimum value of the function: {min_value} at x = {min_x}")
plt.scatter(min_value, min_x, color='red', label='Minimum')
plt.legend(loc='upper right')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Minimum Found by Fibonacci Method')
x_vals = np.linspace(0, 1, 10)
y_vals = np.linspace(0, 3, 10)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)
# Plot the function
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.viridis, alpha=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Minimum Found by Fibonacci Method')
# Add a color bar
cbar = fig.colorbar(ax.plot_surface(X, Y, Z, cmap=cm.viridis), ax=ax, pad=0.1)
cbar.set_label('Function Value')
plt.grid(True)
plt.show()
```
